# Job Posting Notebook

This notebook demonstrates how to use the multi-job-posting-cli to post jobs.

```python
# Example: Load job data and post
import json
from cli.utils.helpers import load_job_file
from adapters.linkedin import LinkedInAdapter
from adapters.workday import WorkdayAdapter

# Load sample job
job_data = load_job_file('examples/sample_job.json')
print(job_data)

# Post to platforms
adapters = {'linkedin': LinkedInAdapter(), 'workday': WorkdayAdapter()}
results = {}
for platform, adapter in adapters.items():
    try:
        response = adapter.post_job(job_data)
        results[platform] = response
        print(f'Success: {platform} - {response}')
    except Exception as e:
        results[platform] = {'error': str(e)}
        print(f'Error: {platform} - {e}')

print('Summary:', results)
```
